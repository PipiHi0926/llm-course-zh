## ä»¥ä¸‹ç¯„ä¾‹æœƒå…ˆä»¥tw_tourism_dfç‚ºä¾‹

```python
""" tw_tourism_df è³‡æ–™å‰è™•ç†(åƒ…ç°¡å–®ç¤ºæ„é€šç”¨æµç¨‹) """
# ç§»é™¤Descriptionæ¬„ä½ç‚ºç©ºå€¼ï¼Œå› ç‚ºæ™¯é»çš„æè¿°ç‚ºå¿…è¦æ¬„ä½è³‡è¨Š
tw_tourism_filtered_df = tw_tourism_df.dropna(subset=['Description'])

# è‹¥nameé‡è¤‡çš„è³‡æ–™ï¼Œå‰‡ç§»é™¤Changetimeæ¯”è¼ƒèˆŠçš„è³‡æ–™
tw_tourism_filtered_df = tw_tourism_filtered_df.sort_values(by='Changetime').drop_duplicates(subset=['Name'], keep='last')

# è³‡æ–™æ¸…ç†å¾Œçš„åŸºæœ¬è³‡è¨Š
tw_tourism_filtered_df.info()
```

```python
# ä¾†æª¢è¦–çœ‹çœ‹tw_tourism_df çš„ä¸€ç­†è³‡æ–™å„æ¬„ä½è³‡è¨Š
tw_tourism_filtered_df.iloc[4]
```

```python
# ä»¥ä¸‹å‡è¨­æˆ‘èªç‚ºName, Description, Region, Town, Orgclass, Keywordæ˜¯è¦ä¿ç•™çš„é‡è¦æ¬„ä½
tw_tourism_filtered_df = tw_tourism_filtered_df[['Name', 'Description', 'Region', 'Town', 'Orgclass', 'Keyword']]

tw_tourism_filtered_df.head(5)
```

```python
import os
import nest_asyncio
import copy

from llama_index.core import Document
from llama_index.core.extractors import (
    SummaryExtractor,
    QuestionsAnsweredExtractor,
    TitleExtractor,
    SummaryExtractor,
    KeywordExtractor,
)
from llama_index.core.node_parser import TokenTextSplitter
from llama_index.core.ingestion import IngestionPipeline
from llama_index.llms.groq import Groq

from pprint import pprint
# é¿å…å·¢ç‹€(åµŒå¥—)å¾ªç’°åˆ¤æ–·å•é¡Œ
nest_asyncio.apply()

api_key = os.getenv("GROQ_API_KEY")

llm = Groq(model="llama3-70b-8192", api_key=api_key)

# ç‚ºäº†ç°¡åŒ–ç¯„ä¾‹ï¼Œæˆ‘åªå…ˆä¿ç•™10ç­†è³‡æ–™
df = tw_tourism_filtered_df.head(10).copy()

# å»ºç«‹ä¸€å€‹LLM model ä¸€æ¨£ by Groq
groq_llm = Groq(model="llama-3.1-70b-versatile", api_key=api_key, temperature=0.1)

# å»ºç«‹ä¸€å€‹ transformations:List ï¼Œè£¡é¢åŒ…å«äº†æ‰€æœ‰çš„æ–‡å­—è™•ç†æµç¨‹
transformations = [
    # TokenTextSplitter(separator=" ", chunk_size=1024, chunk_overlap=64),
    TitleExtractor(nodes=3, llm=groq_llm),
    QuestionsAnsweredExtractor(questions=3, llm=groq_llm),
    SummaryExtractor(summaries=["prev", "self"], llm=groq_llm),
    KeywordExtractor(keywords=5, llm=groq_llm),
]


# åˆ©ç”¨IngestionPipelineä¾†åŸ·è¡Œæ‰€æœ‰çš„å‰›å‰›å®šç¾©transformations
pipeline = IngestionPipeline(transformations=transformations)


df['info'] = df.apply(
    lambda row: ", ".join([
        f"æ™¯é»åç¨±: {row['Name']}",
        f"æ™¯é»ä»‹ç´¹å’Œè³‡è¨Š: {row['Description']}",
        f"åœ°é»åœ¨{row['Region']}{row['Town']}",
        f"ç›¸é—œå±¬æ€§åŒ…å« {row['Orgclass']},{row['Keyword']}"
    ]),
    axis=1
)

documents = [Document(text=text) for text in df['info']]


# åŸ·è¡Œpipelineè™•ç†æ•˜è¿°è³‡æ–™
nodes = pipeline.run(documents=documents)

```
